## 1. My lane as an ML task (type)

## My Lane as an ML Task

**Lane:** Content Refresh

**ML Task Type:** Ranking (Scoring)

The goal is to rank content pages according to their refresh priority. Instead of making a simple yes/no decision, the model assigns each page a score so editors can focus on the highest-priority pages first.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

## Target or Proxy

The starter dataset does not contain a directly observed future outcome showing whether refreshing a page improved its performance.

Therefore, this project uses a **refresh priority score** as a proxy target. The score represents how urgently a page should be refreshed based on historical content and performance features.

If future outcome data becomes available, the proxy can be replaced with an observed target.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

## Success Metric

Since this is a ranking task, the evaluation metric is **Precision@K**.

Precision@K measures how many of the highest-ranked pages are truly valuable refresh candidates. A higher Precision@K indicates better prioritization.

From a business perspective, success means helping editors spend their time on pages with the greatest potential impact.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

## Why ML Beats a Fixed Rule

A fixed rule, such as refreshing every page with low traffic, cannot capture the complex relationships between multiple content and performance signals.

Content performance depends on several interacting factors, including search demand, competition, engagement metrics, and other page characteristics.

Machine learning can combine these signals to produce a more effective ranking than a manually defined rule. The output supports editor decision-making by helping prioritize which pages should be refreshed first.
